## Exercise 2: Multiply List Elements

### Solution 1: Using for loop

````python
# Solution 1: Using for loop
numbers = [1, 2, 3, 4, 5, 6]

for num in numbers:
    print(num * 10)
````

**Output:**
```
10
20
30
40
50
60
```

### Solution 2: Using list comprehension

````python
# Solution 2: Using list comprehension
numbers = [1, 2, 3, 4, 5, 6]
result = [num * 10 for num in numbers]
print(result)
````

**Output:** `[10, 20, 30, 40, 50, 60]`

## Exercise 3: Vector Sum

### Solution using dictionaries

````python
# Read first vector
values = input("Please enter the values of the first vector (x, y, z): ").split()
vector1 = {'x': int(values[0]), 'y': int(values[1]), 'z': int(values[2])}

# Read second vector
values = input("Please enter the values of the second vector (x, y, z): ").split()
vector2 = {'x': int(values[0]), 'y': int(values[1]), 'z': int(values[2])}

# Calculate sum
sum_vector = {
    'x': vector1['x'] + vector2['x'],
    'y': vector1['y'] + vector2['y'],
    'z': vector1['z'] + vector2['z']
}

# Print result
print("The resulting sum vector is:")
print(f"x: {sum_vector['x']}")
print(f"y: {sum_vector['y']}")
print(f"z: {sum_vector['z']}")
````

### Alternative solution using tuples

````python
# Read vectors
x1, y1, z1 = map(int, input("Please enter the values of the first vector (x, y, z): ").split())
x2, y2, z2 = map(int, input("Please enter the values of the second vector (x, y, z): ").split())

# Calculate sum
sum_x = x1 + x2
sum_y = y1 + y2
sum_z = z1 + z2

# Print result
print("The resulting sum vector is:")
print(f"x: {sum_x}")
print(f"y: {sum_y}")
print(f"z: {sum_z}")
````

## Exercise 4: Stock Price Analysis

````python
stock_prices = [150.75, 153.30, 149.50, 155.00, 157.25, 152.80, 154.50]

# 1. Highest price
highest = max(stock_prices)
print(f"Highest price: ${highest:.2f}")

# 2. Lowest price
lowest = min(stock_prices)
print(f"Lowest price: ${lowest:.2f}")

# 3. Average price
average = sum(stock_prices) / len(stock_prices)
print(f"Average price: ${average:.2f}")

# 4. Days above average
days_above_avg = sum(1 for price in stock_prices if price > average)
print(f"Days above average: {days_above_avg}")

# 5. Price change
price_change = stock_prices[-1] - stock_prices[0]
sign = "+" if price_change >= 0 else ""
print(f"Price change: {sign}${price_change:.2f}")
````

**Output:**
```
Highest price: $157.25
Lowest price: $149.50
Average price: $153.30
Days above average: 4
Price change: +$3.75
```

## Exercise 5: Portfolio Manager

````python
def add_stock(portfolio, symbol, quantity):
    """Add stock to portfolio or increase quantity if already exists"""
    if symbol in portfolio:
        portfolio[symbol] += quantity
    else:
        portfolio[symbol] = quantity
    print(f"Added {quantity} shares of {symbol}")

def remove_stock(portfolio, symbol, quantity):
    """Remove stock from portfolio or decrease quantity"""
    if symbol in portfolio:
        if portfolio[symbol] > quantity:
            portfolio[symbol] -= quantity
            print(f"Removed {quantity} shares of {symbol}")
        elif portfolio[symbol] == quantity:
            del portfolio[symbol]
            print(f"Removed all shares of {symbol}")
        else:
            print(f"Error: Only {portfolio[symbol]} shares available")
    else:
        print(f"Error: {symbol} not in portfolio")

def display_portfolio(portfolio):
    """Display all stocks in portfolio"""
    if not portfolio:
        print("Portfolio is empty")
        return
    
    print("Portfolio:")
    for symbol, quantity in portfolio.items():
        print(f"  {symbol}: {quantity} shares")
    print(f"Total stocks: {len(portfolio)}")

# Test
portfolio = {}
add_stock(portfolio, "AAPL", 100)
add_stock(portfolio, "GOOGL", 50)
add_stock(portfolio, "AAPL", 25)
display_portfolio(portfolio)
````

**Output:**
```
Added 100 shares of AAPL
Added 50 shares of GOOGL
Added 25 shares of AAPL
Portfolio:
  AAPL: 125 shares
  GOOGL: 50 shares
Total stocks: 2
```

## Exercise 6: Find Common and Unique Holdings

````python
investor1 = ["AAPL", "GOOGL", "MSFT", "TSLA", "NVDA"]
investor2 = ["AAPL", "AMZN", "MSFT", "META", "NFLX"]

# Convert to sets
set1 = set(investor1)
set2 = set(investor2)

# 1. Common stocks (intersection)
common = set1.intersection(set2)
print("Common stocks:", common)

# 2. All unique stocks (union)
all_stocks = set1.union(set2)
print("All unique stocks:", all_stocks)

# 3. Only Investor 1 (difference)
only_investor1 = set1.difference(set2)
print("Only Investor 1:", only_investor1)

# 4. Exclusive to one (symmetric difference)
exclusive = set1.symmetric_difference(set2)
print("Exclusive to one:", exclusive)
````

**Output:**
```
Common stocks: {'MSFT', 'AAPL'}
All unique stocks: {'NVDA', 'GOOGL', 'MSFT', 'AAPL', 'NFLX', 'META', 'AMZN', 'TSLA'}
Only Investor 1: {'NVDA', 'GOOGL', 'TSLA'}
Exclusive to one: {'NVDA', 'GOOGL', 'NFLX', 'META', 'AMZN', 'TSLA'}
```

## Exercise 7: Stock Information Database

````python
# Create database
stock_database = {
    "AAPL": ("Apple Inc.", "Technology", 150.75),
    "JPM": ("JPMorgan Chase", "Finance", 145.30),
    "GOOGL": ("Alphabet Inc.", "Technology", 138.21),
    "JNJ": ("Johnson & Johnson", "Healthcare", 162.50),
    "MSFT": ("Microsoft Corp.", "Technology", 378.91)
}

# 1. Print all Technology stocks
print("Technology stocks:")
for symbol, (company, sector, price) in stock_database.items():
    if sector == "Technology":
        print(f"  {symbol}: {company} - ${price:.2f}")

# 2. Find most expensive stock
most_expensive_symbol = max(stock_database, key=lambda x: stock_database[x][2])
most_expensive_info = stock_database[most_expensive_symbol]
print(f"\nMost expensive: {most_expensive_symbol} ({most_expensive_info[0]}) - ${most_expensive_info[2]:.2f}")

# 3. Calculate average price
prices = [info[2] for info in stock_database.values()]
average_price = sum(prices) / len(prices)
print(f"Average price: ${average_price:.2f}")
````

**Output:**
```
Technology stocks:
  AAPL: Apple Inc. - $150.75
  GOOGL: Alphabet Inc. - $138.21
  MSFT: Microsoft Corp. - $378.91

Most expensive: MSFT (Microsoft Corp.) - $378.91
Average price: $195.13
```

### Exercise 8a: Round prices

````python
prices = [150.75, 153.30, 149.50, 155.00, 157.25]
rounded_prices = [round(price) for price in prices]
print(rounded_prices)
````

### Exercise 8b: Filter high prices

````python
prices = [150.75, 153.30, 149.50, 155.00, 157.25]
high_prices = [price for price in prices if price > 152]
print(high_prices)
````

### Exercise 8c: Calculate percentage changes

````python
prices = [150.75, 153.30, 149.50, 155.00, 157.25]
first_price = prices[0]
percentage_changes = [((price - first_price) / first_price) * 100 for price in prices]
print(percentage_changes)

# With formatting
formatted_changes = [f"{change:.2f}%" for change in percentage_changes]
print(formatted_changes)
````

**Output:** 

### Exercise 6: Using set operators

````python
# Alternative syntax using operators
investor1 = ["AAPL", "GOOGL", "MSFT", "TSLA", "NVDA"]
investor2 = ["AAPL", "AMZN", "MSFT", "META", "NFLX"]

set1 = set(investor1)
set2 = set(investor2)

common = set1 & set2           # intersection
all_stocks = set1 | set2       # union
only_investor1 = set1 - set2   # difference
exclusive = set1 ^ set2        # symmetric difference

print("Common:", common)
print("All:", all_stocks)
print("Only 1:", only_investor1)
print("Exclusive:", exclusive)
````

### Exercise 8: Nested list comprehension

````python
# Create a 2D list of prices over multiple days
prices_matrix = [[150, 152, 155], [148, 151, 153], [155, 157, 160]]

# Flatten and get all prices above 152
high_prices = [price for day in prices_matrix for price in day if price > 152]
print(high_prices)  # [155, 153, 155, 157, 160]
````

## Bonus: Advanced Techniques

### Dictionary Comprehension

````python
# Create a dictionary from two lists
symbols = ["AAPL", "GOOGL", "MSFT"]
prices = [150.75, 138.21, 378.91]

stock_dict = {symbol: price for symbol, price in zip(symbols, prices)}
print(stock_dict)

# Filter dictionary by value
high_value_stocks = {symbol: price for symbol, price in stock_dict.items() if price > 150}
print(high_value_stocks)
````

### Set Comprehension

````python
# Get unique first letters from stock symbols
symbols = ["AAPL", "AMZN", "GOOGL", "MSFT", "META"]
first_letters = {symbol[0] for symbol in symbols}
print(first_letters)
````

### Using enumerate() and zip()

````python
# enumerate - get index and value
prices = [150.75, 153.30, 149.50]
for day, price in enumerate(prices, start=1):
    print(f"Day {day}: ${price:.2f}")

print()

# zip - combine multiple lists
symbols = ["AAPL", "GOOGL", "MSFT"]
prices = [150.75, 138.21, 378.91]
changes = [2.5, -1.3, 5.2]

for symbol, price, change in zip(symbols, prices, changes):
    print(f"{symbol}: ${price:.2f} ({change:+.2f})")
````
